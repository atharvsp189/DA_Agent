```python
#normalizing columns
df.columns = [col.strip(" ").replace(" ", "_").lower() for col in df.columns]
df.columns
```

```python
# Top/Bottom
def sort_by_rank(col, how_much, colarg=[]):
    if colarg == []:
        return df.sort_values(by=col, ascending=True).head(how_much)
    else:
        return df.sort_values(by=col, ascending=True).head(how_much)[colarg]

```

```python
sort_by_rank('gross_storage_(mcum)', 7, ["धरणाचे_नाव_", "_village_"])
```

```python
def compare_query(query_string, limit = -1):
    if limit != -1:
        return df.query(query_string).head(limit)
    
    return df.query(query_string)
```

```python
col_name = "Latitude"
query_string = f"""{col_name} > 10"""
filter_df = compare_query(query_string)
filter_df
```
